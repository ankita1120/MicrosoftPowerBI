Suppose you are a data analyst working for a movie studio. Your boss has asked you to analyze a dataset containing information about movies, including the movie title, release year, director, genre, and box office revenue. Your boss wants to know:

1.What is the average box office revenue per year for the movies in the dataset?
2.Which genres are the most popular and how many movies have been released in each genre?
3.Who are the top 5 directors in terms of box office revenue?
4.Use Pandas to analyze the dataset and provide a report summarizing your findings. Be sure to include any data cleaning or transformation steps you took to arrive at your conclusions.

```python
import pandas as pd
import google.colab
google.colab.drive.mount('/content/drive')
path = '/content/drive/My Drive/movies.csv'
movies_df = pd.read_csv(path)
movies_df
```

```python
# Calculate average revenue per  year
movies_df['Release Year'] = pd.to_datetime(movies_df['Release Year'], format='%Y') # Added format='%Y' to handle just the year
movies_df['Release Year'] = movies_df['Release Year'].dt.year
# Changed 'Box Office Revenue' to the correct column name 'Box Office'
average_revenue_per_year = movies_df.groupby('Release Year')['Box Office'].mean()
print(f'average_revenue_per_year:\n{average_revenue_per_year}')
```

```python
# Count number of movies released per genre
movies_df['Genre'] = movies_df['Genre'].astype(str).str.split(',')
movies_df = movies_df.explode('Genre')
movies_df['Genre'] = movies_df['Genre'].astype(str).str.strip()
genre_counts = movies_df['Genre'].value_counts()
print(f'genre_counts:\n{genre_counts}')
```

```python
# Calculate top directors by revenue
movies_df['Director'] = movies_df['Director'].astype(str).str.split(',')
movies_df = movies_df.explode('Director')
movies_df['Director'] = movies_df['Director'].astype(str).str.strip()
top_directors = movies_df.groupby('Director')['Box Office'].sum().nlargest(5)
print(f'top_directors:\n{top_directors}')
```
