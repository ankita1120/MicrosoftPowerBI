```python
import pandas as pd
import numpy as np

# Assuming the file path is correct
file_path = '/content/drive/My Drive/imdb_data.csv'
imdb_data = pd.read_csv(file_path)

# Function to capitalize titles
def to_uppercase(column):
  return column.upper()

# Apply the function to create 'Capitalized_title' column
imdb_data['Capitalized_title'] = imdb_data['title'].apply(to_uppercase)

# Function to calculate revised profit
def revised_profit(revenue, budget):
  if revenue > 0:
    new_profit = revenue - budget
  else:
    new_profit = np.nan
  return new_profit

# Apply the function to create 'new_profit' column
imdb_data['new_profit'] = imdb_data.apply(lambda x: revised_profit(x['revenue'], x['budget']), axis=1)

# Create three DataFrames based on runtime
imdb1 = imdb_data[imdb_data['runtime'] >= 150]
imdb2 = imdb_data[(imdb_data['runtime'] < 150) & (imdb_data['runtime'] > 100)]
imdb3 = imdb_data[imdb_data['runtime'] <= 100]

# Create a new DataFrame with the selected columns
new_data = imdb_data[['new_profit', 'Capitalized_title']].head()

# Display the new DataFrame
print(new_data)
```

```python
import pandas as pd

# Assuming 'imdb_data' is your DataFrame
# and you want to save it as 'imdb_data.csv'
imdb_data.to_csv('imdb_data.csv', index=False)

# Now you can read it back using pd.read_csv
imdb_df = pd.read_csv('imdb_data.csv')


imdb_df.columns
imdb_df.head(2)


# Convert the "Release Date" column to datetime type if it's not already in datetime format
imdb_df['release_date'] = pd.to_datetime(imdb_df['release_date'], errors='coerce')

# Extract the release year and add it as a new column
imdb_df['release_year'] = imdb_df['release_date'].dt.year

# We can find the maximum movie runtime across different years
max_runtime_movies_by_year = imdb_df.groupby('release_year')['runtime'].max().reset_index()

max_runtime_movies_by_year.rename(columns = {'runtime' : 'max_runtime'},inplace=True)
max_runtime_movies_by_year.head()

# We can find the median movie runtime across different years
imdb_df.groupby('release_year')['runtime'].median().reset_index()
```

```python
# Group the data frame by year and item and extract a number of stats from each group
imdb_df.groupby(['release_year', 'original_language'],as_index=False).agg({'runtime':"mean",'revenue': "sum",'imdb_id':"count"}).sort_values(by='release_year').rename(columns ={'runtime':'avg_runtime','revenue':'total_revenue','imdb_id':'num_movies'})
```

```python
import pandas as pd
imdb_df['profit'] = imdb_df['revenue'] - imdb_df['budget']
imdb_df['profit_margin'] = imdb_df['profit'] / imdb_df['revenue']
imdb_df['is_profitable'] = imdb_df['profit'].apply(lambda x: 'Yes' if x > 0 else 'No')
print(imdb_df.head(10))
```

Your friend is a film enthusiast and has decided to explore movies in languages other than English. He wants recommendations on which movies to watch next.

Using the dataset provided, your task is to find out the language with the highest mean popularity rating and then list the top 5 movies in that language.

Can you help your fellow film enthusiast friend by completing this task and providing them with recommendations for non-English movies?



```python
import pandas as pd

# Calculate the mean popularity for each language
language_popularity = imdb_df.groupby('original_language')['popularity'].mean()

# Find the language with the highest mean popularity
highest_popularity_language = language_popularity.idxmax()

# Filter the DataFrame to include only movies in the highest popularity language
filtered_movies = imdb_df[imdb_df['original_language'] == highest_popularity_language]

# Sort the filtered movies by popularity in descending order
sorted_movies = filtered_movies.sort_values(by='popularity', ascending=False)

# Get the top 5 movies
top_5_movies = sorted_movies.head(5)

# Print the results
print(f"The language with the highest mean popularity rating is: {highest_popularity_language}")
print("\nTop 5 movies in that language:")
print(top_5_movies[['title', 'popularity']])
```
